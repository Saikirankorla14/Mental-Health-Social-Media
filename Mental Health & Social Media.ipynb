```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download necessary NLTK data
try:
    nltk.data.find('tokenizers/punkt')
except nltk.downloader.DownloadError:
    nltk.download('punkt')
try:
    nltk.data.find('corpora/stopwords')
except nltk.downloader.DownloadError:
    nltk.download('stopwords')
try:
    nltk.data.find('corpora/wordnet')
except nltk.downloader.DownloadError:
    nltk.download('wordnet')


stop_words = set(stopwords.words('english'))
# Using Porter Stemmer as it's simpler and often sufficient for many tasks
stemmer = PorterStemmer()
# Using WordNet Lemmatizer for more accurate base forms
lemmatizer = WordNetLemmatizer()


def process_text(text):
    tokens = word_tokenize(text) # 3. Tokenize the text
    filtered_tokens = [word for word in tokens if word not in stop_words] # 4. Remove stop words
    # Applying stemming or lemmatization (choose one based on preference/task)
    # Here, we will use lemmatization as it usually provides better results
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens] # 5. Apply lemmatization
    return ' '.join(lemmatized_tokens) # 6. Join tokens back into a string

df['processed_text'] = df['cleaned_text'].apply(process_text)
display(df[['text', 'cleaned_text', 'processed_text']].head())
```

```python
import re

def clean_text(text):
    text = text.lower() # 1. Convert to lowercase
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE) # 2. Remove URLs
    text = re.sub(r'@\w+', '', text) # Remove mentions
    text = re.sub(r'#\w+', '', text) # Remove hashtags
    text = re.sub(r'[^a-z0-9\s]', '', text) # Remove special characters
    return text

df['cleaned_text'] = df['text'].apply(clean_text)
display(df[['text', 'cleaned_text']].head())
```
